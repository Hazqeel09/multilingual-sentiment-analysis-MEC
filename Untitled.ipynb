```python
from bs4 import BeautifulSoup
import requests

URL = 'https://www.popularonline.com.my/default/malay.html?did=1996'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
containers = soup.findAll('div',{'class':'col-sm-3'})
```

```python
for container in containers:
    print(container.get_text().strip().partition("\n")[2].partition("\n")[2].partition("\n")[2][34:-41].strip())
    print("---------------")
```
