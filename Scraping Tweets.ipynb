```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
url = "https://mobile.twitter.com/khairulaming?lang=en"

for page in range(0,5):
    res = Request(url.format(page))
    page = urlopen(res)
    soup = BeautifulSoup(page,'html.parser')
    tweets = soup.find_all(attrs={"class":'tweet-text'})
    for tweet in tweets:
        print(tweet.text.strip())
```
